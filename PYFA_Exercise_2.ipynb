```python
# 3.1 В каком диапазоне изменяются стоимости недвижимости?
answ = 'Стоимость домов меняется в диапазоне от ' + f_format(round(df['price'].min()/1000,1), num) + ' тыс. до ' + f_format( round(df['price'].max()/1000,1), num) + ' тыс.'
answ
```

```python
# 3.2 Какую долю в среднем занимают жилая площадь от всей площади по всем домам?

total_sqware = df['sqft_lot'].sum()
living_sqware = df['sqft_living'].sum()
living_prorate = living_sqware / total_sqware

'Доля жилой площади в данных составляет: ' + f_format(living_prorate, perc)
```

```python
# 3.3 Как много домов с разными этажами в данных?
df['floors'].value_counts()
```

```python
# 3.4 Насколько хорошие состояния у домов в данных?
df['condition'].value_counts()
```

```python
# 3.5 Найдите года, когда построили первый дом, когда построили последний дом в данных?
year_first = df['yr_built'].min()
year_last = df['yr_built'].max()

"Первый год - " + str(year_first) + ", последний год -" + str(year_last)
```

###Задание №4
------------------------

```python
# 4.1 Сколько в среднем стоят дома, у которых 2 спальни?

two_bedrooms_price = df[round(df['bedrooms'],0) == 2]['price'].mean()
'Средняя стоимость дома составляет: ' + f_format(two_bedrooms_price, num)
```

```python
# 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?
avg_price = df[df['price']>600000]['sqft_lot'].mean()
'Средняя площадь S составляет: ' + f_format(avg_price, num)
```

```python
# 4.3 Как много домов коснулся ремонт?
qty = df[df['yr_renovated']!=0]['yr_renovated'].count()
f_format(qty, num)
```

```python
# 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?
price_gr10 = df[df['grade']>10]['price'].mean()
price_gr4 = df[df['grade']<4]['price'].mean()
diff_gr10vs4 = round(price_gr10 - price_gr4,0)

'Отличие в среднем составляет: ' + f_format(diff_gr10vs4, num)
```

```python
# 5.1 Выберите дом клиенту: Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом.
#Сколько вариантов есть у клиента?

dream_house = df[ (df['waterfront']!=0) & (df['bathrooms']>=3) & (df['sqft_basement']!=0)]['id'].count()
'Всего вариантов для клиента: ' + f_format(dream_house, num)
```

```python
# 5.2 Выберите дом клиенту: Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года.
# В какой ценовом диапазоне будут дома?

#dream_house_price = df[ ((df['waterfront']==1) | (df['view']==4)) & (df['condition']==5) & (df['yr_built']>=1980) ]['price']

# Рещение с помощью языка запросов
waterfront_filter = 1
view_filter = 4
condition_filter = 5
year_built_filter = 1980

dream_house_price =df.query( "(waterfront== @waterfront_filter | view == @view_filter) & condition==@condition_filter  & yr_built >= @year_built_filter")['price']

'Ценовой диапазон составляет: ' + f_format(round(dream_house_price.min()/1000,0), num) + ' тыс. до ' + f_format(round(dream_house_price.max()/1000,0),num) + ' тыс.'
```

```python
# 5.3 Выберите дом клиенту: Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

avg_condition = df[ (df['sqft_basement']==0)  &  (df['floors']==2)  & (df['price']<150000) ]['condition'].mean()
'{:.1f}'.format(avg_condition)
```
